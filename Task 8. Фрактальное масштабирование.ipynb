На электронном заводе в городе M выпускают микросхемы, которые затем могут быть использованы в качестве вычислительных мощностей в различных электронных приборах. Считается, что мощность микросхемы равна ее площади.

Первое поколение микросхем было в форме единичного квадрата, затем каждое поколение завод наращивал мощности микросхем добавляя на каждую сторону микросхемы предыдущего поколения по участку в форме прямоугольного треугольника, как показано на картинке ниже.




Определите, чему равна мощность микросхемы n-го поколения. Дайте ответ с точностью до 
10^-6. Считать, что мощность первого поколения микросхем равна 1.

 

Входные данные

В единственной строке входных данных содержится целое число n (2≤n≤10^9).

 

Выходные данные

Выведите единственное число — мощность микросхемы n-го поколения. Ваш ответ будет засчитан, если он отличается от ответа жюри не более чем на 
10^−6.

Примечание. Вывести в ответе все 6 знаков после запятой, даже если все 0! 

## Финальное решение

```python
"""
По факту задача решена математически
"""

def hock(n):
    s = 0
    for i in range(0, n - 2 + 1):
        s += 2 ** (i // 2 - i)
    return s - 2 + 1 / 2 ** (n - 2)


def find_square(n: int) -> float:
    s_t = 1 / 4
    s_k = 1 / 4
    m = (n - 5) // 2 + (1 if (n - 5) % 2 == 0 else 2)
    if n <= 3:
        return n
    else:
        if n % 2 != 0:
            return 3 + 8 * (
                s_k * (1 - 1 / 2 ** ((n - 3) / 2)) + s_t * hock(n)
            )
        else: 
            return 3 + 8 * (
                s_k * (1 - 1 / 2 ** (m - 1) + 1 / 2 ** (m + 1))
                + s_t * (1 - 1 / 2 ** ((n - 2) / 2)) ** 2
            )


a = find_square(int(input()))
print(f"{{:.6f}}".format(round(a, 6)))
```
