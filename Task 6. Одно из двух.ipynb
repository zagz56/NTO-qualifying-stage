```python
def hex_to_binary(hex_byte):
    """
    Преобразование в двоичное представление
    """
    # Преобразование шестнадцатеричного числа в бинарное представление
    binary_string = bin(int(hex_byte, 16))[2:].zfill(
        8
    )  # Представляем число в виде строки битов, дополняя нулями до 8 символов

    # Распределение битов по переменным от a0 до a7
    a0, a1, a2, a3, a4, a5, a6, a7 = map(
        int, binary_string
    )  # Преобразование строкового представления в целые числа

    return a0, a1, a2, a3, a4, a5, a6, a7


def full_adder(A1, A2, B1, B2, P0):
    """
    2-разрядный полный сумматор
    """
    # XOR для первого разряда
    S2 = A1 ^ B1 ^ P0
    # Вычисление переноса на следующий разряд
    P1 = (A1 & B1) | (P0 & (A1 ^ B1))

    # XOR для второго разряда
    S1 = A2 ^ B2 ^ P1
    # Вычисление переноса на третий разряд
    P2 = (A2 & B2) | (P1 & (A2 ^ B2))

    return S1, S2, P2


def logical_nand(S1, S2, P2, a4):
    """
    Логическое и-не
    """
    return 0 if S1 and S2 and P2 and a4 else 1


number_1 = input("")
number_2 = input("")

# number_1 = "A9"
# number_2 = "FC"

a0, a1, a2, a3, a4, a5, a6, a7 = hex_to_binary(number_1)
P2 = 0

A1 = a7
B1 = a6
A2 = a5
B2 = a5
P0 = P2

S1, S2, P2 = full_adder(A1, A2, B1, B2, P0)

a0, a1, a2, a3, a4, a5, a6, a7 = hex_to_binary(number_2)

А1 = a7
B1 = a6
A2 = a5
B2 = a5
P0 = P2

S1, S2, P2 = full_adder(A1, A2, B1, B2, P0)

out_of_logical_nand = logical_nand(S1, S2, P2, a4)

if out_of_logical_nand:
    print("LED OFF")
else:
    print("LED ON")
```
