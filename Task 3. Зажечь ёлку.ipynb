```python
import random

random.seed(42)


def generate_random_list(m, n):
    chars = ["x", "o", "."]
    random_list = []

    for _ in range(m):
        random_string = "".join(random.choice(chars) for _ in range(n))
        random_list.append(random_string)

    return random_list


m = 14  # количество строк
n = 14  # количество символов в строке
t = 1

# switches = ['.o.', 'oxo']

switches = generate_random_list(m, n)
switches = tuple(switches)
for row in switches:
    print(row)
```
